```python
import os
import shutil

DATASET_PATH = r"C:\Users\Rakshitha\Military\military_object_dataset"
REDUCED_PATH = r"C:\Military_Project\reduced_dataset_4class"

print("Paths defined successfully ")
```

```python
class_map = {
    0: 0,
    1: 1,
    2: 2,
    3: 3
}
```

```python

for split in ["train", "val", "test"]:

    os.makedirs(f"{REDUCED_PATH}/{split}/images", exist_ok=True)
    os.makedirs(f"{REDUCED_PATH}/{split}/labels", exist_ok=True)

    img_folder = os.path.join(DATASET_PATH, split, "images")
    label_folder = os.path.join(DATASET_PATH, split, "labels")

    for file in os.listdir(label_folder):

        new_lines = []

        with open(os.path.join(label_folder, file)) as f:
            lines = f.readlines()

        for line in lines:
            cls = int(line.split()[0])

            if cls in class_map:
                new_lines.append(line)

        if new_lines:
            img_name = file.replace(".txt", ".jpg")
            src_img = os.path.join(img_folder, img_name)

            if os.path.exists(src_img):
                shutil.copy(src_img,
                            os.path.join(REDUCED_PATH, split, "images"))

            with open(os.path.join(REDUCED_PATH, split, "labels", file), "w") as f:
                f.writelines(new_lines)

print("4-class dataset created successfully âœ…ðŸ”¥")
```

```python
model = YOLO("yolov8n.pt")

model.train(
    data="military_4class.yaml",
    epochs=40,
    imgsz=640,
    batch=4,       
    patience=5,  
    name="military_4class_yolo"
)
```

```python
metrics = model.val(data="military_4class.yaml")

```
